```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mbtc = pd.read_csv('data.csv')
```

```python
mbtc.columns
```

```python
mbtc.info()
```

```python
# we want to turn date from a object to date time
mbtc['date'] = pd.to_datetime(mbtc['date'])
```

```python
# now we can convert date to the index
mbtc.set_index('date')
```

```python
# lets graph btc and CFGI
plt.plot(mbtc['price'], color ='black')
plt.plot(mbtc['cfgi'], color ='red')
```

```python
# Looks like we will need twin axiss ^ and also set the index
```

```python
import matplotlib.pyplot as plt

# Create the figure and the first axis
fig, ax1 = plt.subplots()

# Plot BTC prices on the first y-axis (left side)
ax1.plot(mbtc['price'], color='black', label='BTC Price')
ax1.set_xlabel('Date')  # Set the common x-axis label
ax1.set_ylabel('BTC Price', color='black')
ax1.tick_params(axis='y', labelcolor='black')

# Create a second y-axis sharing the same x-axis
ax2 = ax1.twinx()
ax2.plot(mbtc['cfgi'], color='red', label='CFGI')
ax2.set_ylabel('CFGI', color='red')
ax2.tick_params(axis='y', labelcolor='red')

# Optionally, add a title
plt.title("BTC Price and CFGI with Twin Axes")

# Optionally, add legends (requires a bit more work when using twin axes)
lines_1, labels_1 = ax1.get_legend_handles_labels()
lines_2, labels_2 = ax2.get_legend_handles_labels()
ax1.legend(lines_1 + lines_2, labels_1 + labels_2, loc='upper left')

# Display the plot
plt.show()
```
